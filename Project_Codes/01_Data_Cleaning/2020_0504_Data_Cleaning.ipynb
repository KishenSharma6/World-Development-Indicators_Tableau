# Data Cleaning

## Drop irrelevant columns

```python
#Drop unnecessary cols 
df.drop(['Indicator Code','Country Code','Unnamed: 64'], axis = 1, inplace = True)

#Check
display(df.head(3))
```

## Tidy data

**Aggregate years into a single column**

```python
#Gather years
df = pd.melt(df, id_vars=['Country Name','Indicator Name'], var_name='Year')

#Check
display(df.head(2))
```

**Spread Indicator Name into column headers**

```python
#Convert NaN in value column into -1 place holders
df['value'].fillna(-1, inplace=True)

#Groupby to create a unique index. Unstack Indicator name
df = df.groupby(['Country Name', 'Indicator Name', 'Year'])['value'].sum()
df = df.unstack(level = -2)

#Replace -1's with NaN
df.replace(to_replace= -1, value= np.nan, inplace=True)

#Check
display(df.head())
```

# Write cleaned file to CSV

```python
#Print final shape of df
print('Shape of cleaned data:', df.shape)

#Write df to local machine
df.to_csv(path + '/Data/02_Cleaned_Data/2020_0505_Cleaned_WDIData.csv')
```
